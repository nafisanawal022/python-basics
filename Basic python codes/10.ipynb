```python
## Task 3##
class Node:
    
    def __init__(self,element,next):
        self.element= element
        self.next= next
        
    def printNode(self):
        print(self.element, "-", self.next)
        

class MyList():
    def __init__(self,LL=None):
        if (LL==None):
            self.head = None
            self.size = 0
            self.tail = None
        
        elif (type(LL)==list):
            self.size =0
            self.head= None
            self.tail = None
            
            for a in LL:
                n = Node(a,None)
                
                if (self.head==None):
                    self.size=self.size+1
                    self.head = n
                    self.tail=n
                    
                else:
                    self.size = self.size+1
                    self.tail.next = n
                    self.tail = n
                    
        else:
            self.head = None
            self.tail = None
            self.size = LL.size
            n=LL.head
            
            while (n!=None):
                n2 = Node(n.element, None)
                
                if (self.head == None):
                    self.head = n2
                    self.tail = n2
                    
                else:
                    self.tail.next = n2
                    self.tail = n2
                    
                n = n.next
                
    def printList(self):
        
        if (self.size==0):
            print ("Empty List")
            
        else:
            n = self.head
            while (n!=None):
                n.printNode()
                n=n.next
                
                    
    def evenList(self):
        a=self.head
        while(a!=None):
            if(a.element%2==0):
                print(a.element)
            a=a.next
            
    def search(self,elem):
        n=self.head
        count=0
        while(n!=None):
            if(n.element==elem):
                count=count+1
            n=n.next
            
        if(count>0):
            return True
        else:
            return False
        
    def reverse(self):
        before=None
        present=self.head
        
        while(present!=None):
            after = present.next
            present.next=before
            before=present
            present=after
        self.head=before
        
    def sort(self):
        n=self.head
        min=None
      
        if(self.head==None):
            return self.head
        else:
            while(n!=None):
                min=n.next
                while(min!=None):
                    if(n.element>min.element):
                        store=n.element
                        n.element=min.element
                        min.element=store
                    min=min.next
                n=n.next
    
    def sum(self):
        n=self.head
        sum=0
        while(n!=None):
            sum=sum+n.element
            n=n.next
        return sum
    
    def rotate(self,direction,k):
        self.direction=direction
        self.k=k
        if (self.direction=="left"):
            i=0
            while (i<k):
                oldHead=self.head
                self.head=self.head.next
                tail=self.head
                while(tail.next!=None):
                    tail=tail.next
                tail.next=oldHead
                oldHead.next=None
                
                i=i+1
                
        elif (self.direction=="right"):
            count=0
            n=self.head
            while(n!=None):
                n=n.next
                count=count+1
            print(count)
            i=0
            while (i<(count-k)):
                oldHead=self.head
                self.head=self.head.next
                tail=self.head
                while(tail.next!=None):
                    tail=tail.next
                tail.next=oldHead
                oldHead.next=None
                
                i=i+1
            
a=MyList([3,2,5,1,8,9])
print("-----------Even List----------")
a.evenList()
a.printList()
print("-----------Search----------")
a.search(7)
a.printList()
print("-----------Reverse----------")
a.reverse()
a.printList()
print("-----------Sort----------")
a.sort()
a.printList()
print("-----------Sum----------")
a.sum()
a.printList()
print("-----------Rotate----------")
a.rotate("left",2)
a.printList()
```
